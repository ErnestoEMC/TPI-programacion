### TABLA DE MULTIPLICAR CON WHILE

```python
n=int(input('ingrese un numero->'))

i=0

while i<10:
  i=i+1

  print(f'{n} x {i} = {n*i}')

```

```python
n=int(input('ingrese un numero->'))

while n<0:
  print('error')
  n=int(input('ingrese un numero positivo'))


print(f'La raiz cuadrada del numero {n} es {round(n**0.5,3)}')



```

```python
edad=int(input('ingrese su edad->'))
while edad <0 or edad>100:
  print('ha ingresado una edad incorrecta')

  edad=int(input('ingrese nuevamente su edad->'))

print(f'finalizo el proceso y su edad es {edad} años')
```

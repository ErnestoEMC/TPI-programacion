### 1.	Crear una tupla simple

```python
tupla = (1, 2, 3)


print(tupla)
```

### Acceder a un elemento de la tupla

```python
tupla = ('a', 'b', 'c')
print(tupla[1])
```

### 3.	Tupla con diferentes tipos de datos

```python
tupla = (1, "hola", 3.5, True)
print(tupla)
```

### Iterar sobre una tupla

```python
tupla = (10, 20, 30)
for elemento in tupla:
    print(elemento)
```

### Longitud de una tupla

```python
tupla = (1, 2, 3, 4)
print(len(tupla))
```

### Contar cuántas veces aparece un elemento

```python
tupla = (1, 2, 2, 3, 2)
print(tupla.count(2))
```

### Encontrar el índice de un elemento

```python
tupla = ('a', 'b', 'c', 'd')
print(tupla.index('c'))
```

### Convertir una lista a tupla

```python
lista = [1, 2, 3]
tupla = tuple(lista)
print(tupla)
```

### Convertir una tupla a lista


```python
tupla = (1, 2, 3)
lista = list(tupla)
print(lista)
```

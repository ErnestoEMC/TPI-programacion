### uso de continue

```python
for num in range(10):

  if num%2==0:
    continue

  print(num)
```

```python
num=0
while num <10:

  num+=1
  if num%2==0:
    continue

  print(num)



```

### USO DE BREAK

```python
for num in range(10):
  if num==5:
    break

  print(num)
```

```python
num=0

while num<10:
  num=num+1
  if num == 5:
    break

  print(num)
```

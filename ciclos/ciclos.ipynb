### programa que valide si una contraseña es segura
- tiene mas de 8 caracteres
- tiene al menos 1 letra mayuscula
- tiene al menos un numero

```python


def validar(cadena):

  respuesta1=False
  respuesta2=False
  respuesta3=False


  for i in cadena:

    if i.isnumeric():

      respuesta1=True

    if i.isupper():
      respuesta2=True

    if len(cadena)>8:
      respuesta3=True

  if respuesta1 and respuesta2 and respuesta3:
    print('La contraseña es segura')
  else:
    print('La contraseña no es segura')

cadena=input('Ingrese una contraseña: ')

validar(cadena)






```

### programa que valide si un numero es primo

```python
def primo(n):

  conta=0
  for i in range(1,n+1):
    if n%i==0:
      conta+=1

  if conta==2:
    print('El numero es primo')
  else:
    print('El numero no es primo')

primo(13)






```

### Programa que calcula el nuevo salario de un trabajador , si obtuvo un incremento de x%.

```python
def salario(salario,x):

  nuevo=salario+salario*(x/100)

  return nuevo

sala=float(input('Ingrese el salario: '))
xp=float(input('Ingrese el porcentaje de incremento: '))


resp=salario(sala,xp)

print(f'El nuevo salario es {resp:.2f}')

```

### programa que pida 2 numeros y realice las siguientes operaciones:
###-suma
###-resta
###-multiplicacion
###-division
###-exponenciacion
###-radicacion

```python
import os


def suma(a,b):
  return a+b

def resta(a,b):
  return a-b

def multiplicacion(a,b):
  return a*b

def division(a,b):
  return a/b

def potencia(a,b):
  return a**b

def raiz(a,b):
  return a**(1/b)


print('=======================================')
print('Bienvenido a la calculadora')
print('=======================================')
while True:
  os.system('cls')

  print('1.suma')
  print('2.resta')
  print('3.multiplicacion')
  print('4.division')
  print('5.potencia')
  print('6.raiz')
  print('7.salir')

  a=float(input('Ingrese el primer numero: '))
  b=float(input('Ingrese el segundo numero: '))

  opcion=int(input('Ingrese una opcion: '))

  if opcion==1:

    print(f'La suma de {a} y {b} es {suma(a,b)}')
  elif opcion==2:
    print(f'La resta de {a} y {b} es {resta(a,b)}')
  elif opcion==3:
    print(f'La multiplicacion de {a} y {b} es {multiplicacion(a,b)}')

  elif opcion==4:
    print(f'La division de {a} y {b} es {division(a,b)}')

  elif opcion==5:
    print(f'La potencia de {a} y {b} es {potencia(a,b)}')

  elif opcion==6:

    print(f'La raiz de {a} y {b} es {raiz(a,b)}')
  elif opcion==7:

    print('Gracias por usar la calculadora')
    break

```

### Escribir una función que reciba un número entero positivo y devuelva su factorial.

```python
def facto(n):

  fact=1

  for i in range(1,n+1):

    fact=fact*i

  return fact

n=int(input('ingrese un numero positivo: '))

print(f'el factorial del numero {n} es {facto(n)}')
```

### Escribir una función que calcule el área de un círculo y otra que calcule el volumen de un cilindro usando la primera función.

```python
import math

def circulo(radio):
  area=math.pi*radio**2
  return area

def cilindro(radio,altura):

  volumen=circulo(radio)*altura

  return volumen

radio=float(input('Ingrese el radio del circulo: '))
altura=float(input('Ingrese la altura del cilindro: '))

print(f'El area del circulo es {circulo(radio)}')

print(f'El volumen del cilindro es {cilindro(radio,altura)}')

```

### Escribir una función que reciba una muestra de números en una lista y devuelva su media.

```python
def promedio(lista):
  suma=0
  for i in lista:

    suma+=i

  media=suma/len(lista)
  return media

resultado=promedio('[3,5,4,5]')

print(f'La media es {resultado}')


```

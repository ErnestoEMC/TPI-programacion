### Escribir un programa que pregunte al usuario su nombre, edad, dirección y teléfono y lo guarde en un diccionario. Después debe mostrar por pantalla el mensaje <nombre> tiene <edad> años, vive en <dirección> y su número de teléfono es <teléfono>.


```python
nombre=input('ingrese su nombre : ')
edad= int(input('ingrese su edad : '))
direccion=input('ingrese su direccion : ')
telefono= int(input('ingrese su telefono : '))

datos={'nombre':nombre,'edad':edad,'direccion':direccion, 'telefono':telefono}

print(f"{datos['nombre']} tiene {datos['edad']} años , vive en {datos['direccion']} y su telefono es {datos['telefono']}")
```

### Escribir un programa que guarde en un diccionario los precios de las frutas de la tabla, pregunte al usuario por una fruta, un número de kilos y muestre por pantalla el precio de ese número de kilos de fruta. Si la fruta no está en el diccionario debe mostrar un mensaje informando de ello.

### Manzana	0.80
### Pera	0.85
### Naranja	0.70
### Platano 2.50

```python
frutas={'Manzana':0.80,
        'Platano':1.35,
      'Pera':0.85,
      'Naranja':0.70

      }
compra= input('Que fruta desea comprar: ')

if compra in frutas:

  kilos= int(input('Cuantos kilos desea comprar: '))

  print(f'El precio total a pagar por la {compra} es {frutas[compra]*kilos} soles')

else:
  print('la fruta no esta en el diccionario')
```

### Escribir un programa que pregunte una fecha en formato dd/mm/aaaa y muestre por pantalla la misma fecha en formato dd de <mes> de aaaa donde <mes> es el nombre del mes.

```python

meses={'1':'Enero',
       '2':'Febrero',
       '3':'Marzo',
       '4':'Abril',
       '5':'Mayo',
       '6':'Junio',
       '7':'Julio',
       '8':'Agosto',
       '9':'Septiembre',
       '10':'Octubre',
       '11':'Noviembre',
       '12':'Diciembre'}

fecha=input('ingrese una fecha con el formato dd/mes/año : ')

indice=fecha.split('/')


print(f'La fecha ingresada es {indice[0]}/{meses[indice[1]]}/{indice[2]}')
```

### Escribir un programa que almacene el diccionario con los créditos de las asignaturas de un curso {'Matemáticas': 6, 'Física': 4, 'Química': 5} y después muestre por pantalla los créditos de cada asignatura en el formato <asignatura> tiene <créditos> créditos, donde <asignatura> es cada una de las asignaturas del curso, y <créditos> son sus créditos. Al final debe mostrar también el número total de créditos del curso.

```python
cursos={'Matemáticas': 6, 'Física': 4, 'Química': 5}

total=0

for curso in cursos:
  print(f'{curso} tiene {cursos[curso]} creditos')
  total= total+cursos[curso]

print(f'el total de creditos del curso es {total}')
```

### creacion y acceso a un diccionario


```python
estudiantes = {
    "Ana": 20,
    "Pedro": 22,
    "María": 19,
    "Juan": 21
}


print(estudiantes)

print(f"La edad de María es {estudiantes['María']} años")

print(f"la edad de Ana es {estudiantes['Ana']} años")

print(f"la edad de Juan es {estudiantes['Juan']} años")

```

### Modificación de valores en un diccionario

```python
productos = {
    "manzana": 1.2,
    "plátano": 0.8,
    "naranja": 1.0,
    "uva": 2.5
}

productos["naranja"] = 4.5

print(productos)
```

### Eliminación de elementos en un diccionario

```python
ciudades = {
    "Madrid": 3223000,
    "Barcelona": 1620000,
    "Valencia": 794000,
    "Sevilla": 689000
}

print(ciudades)

del ciudades["Sevilla"]

print(ciudades)
```

### Recorrido de diccionario


```python
frutas = {
    "manzana": "rojo",
    "plátano": "amarillo",
    "naranja": "naranja",
    "uva": "morado"
}

for fruta,color in frutas.items():

  print(f" la fruta {fruta} es de color {color}")

```

### Crear diccionario a partir de listas

```python
nombres = ["Ana", "Pedro", "María", "Juan"]
edades = [20, 22, 19, 21]
estudiantes = {}

for i in range(4):

  estudiantes[nombres[i]]=edades[i]

print(estudiantes)
```

### filtrado de diccionarios

```python
edades = {
    "Ana": 25,
    "Pedro": 17,
    "María": 30,
    "Juan": 16,
    "Lucía": 22
}

for nombre, edad in edades.items():

  if edad >=18:

    print(f' {nombre} es mayor de edad')

  else:
    print(f' {nombre} es menor de edad')

```

### agregar contacto

```python


datos = {
    "Ana": {
        "telefono": "123456789",
        "email": "ana@example.com",
        "direccion": "Calle Principal 123"
    },
    "Pedro": {
        "telefono": "987654321",
        "email": "pedro@example.com",
        "direccion": "Avenida Central 456"
    }

}


def agregar_contacto(datos, nombre, telefono, email, direccion):
    datos[nombre] = {
        "telefono": telefono,
        "email": email,
        "direccion": direccion
    }


agregar_contacto(datos, 'Ana', '123456789', 'ana@example.com', 'calle principal 123')

print(agregar_contacto)
```

### Métodos get() y setdefault()

```python
inventario = {
    "manzanas": 50,
    "plátanos": 30,
    "naranjas": 40,
    "uvas":25
}

cantidad_manzanas = inventario.get("manzanas", 50)
print(f"Cantidad de manzanas (con get): {cantidad_manzanas}")

inventario.setdefault("mango", 40)
print(f"Inventario después de setdefault con mango: {inventario}")


```

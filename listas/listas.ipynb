### Divisores de un numero

```python
num=int(input('ingrese un numero entero->'))

for i in range(1,num+1):
  if num%i==0:

    print (f'el numero {i} es divisor de {num}')

```

### Numeros primos

```python
num=int(input('ingrese un numero entero->'))
contador=0

for i in range(1,num+1):
  if num%i==0:
    contador=contador+1

if contador==2:

  print(f'el numero {num} es primo')

else:
  print(f'el numero {num} no es primo')
```

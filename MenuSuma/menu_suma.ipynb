### 1. Hacer un programa que tome las 3 notas de un estudiante . Diga la nota final del curso
debe tener en cuenta que la primera y segunda nota vale el 30% y la ultima nota vale 40%.

```python
def promedio(n1,n2,n3):
  nota=n1*0.3+n2*0.3+n3*0.4
  return nota
n1=int(input('ingrese la nota 1: '))
n2=int(input('ingrese la nota 2; '))
n3=int(input('ingrese la nota 3: '))

promedio(n1,n2,n3)

print(f'La nota final es{promedio(n1,n2,n3)}')
```
